```python
test_song_hist = test_data['song_id'].value_counts()
print(test_song_hist)
```

```python
new_song_count = 0
new_song_accumulate = 0
less_song_count = 0
less_song_accumulate = 0
total_test_songs = 0
total_test_songs_accumulate = 0
for song_id, count_num in test_song_hist.iteritems():
    train_song_count = song_hist.get(song_id, default=-1)
    total_test_songs += 1
    total_test_songs_accumulate += count_num
    if train_song_count < 0:
        new_song_accumulate += count_num
        new_song_count += 1
    elif train_song_count < 5:
        less_song_count += 1
        less_song_accumulate += count_num
        # print('song:\"{}\" count {} times in test, count {} times in train'.format(song_id, count_num, train_song_count))
print('there are {}(total {}) songs in test dataset'.format(total_test_songs, total_test_songs_accumulate))
print('{}/{} of them are count less than 5 in train dataset'.format(less_song_count, less_song_accumulate))
print('{}/{} of them never exists in train dataset'.format(new_song_count, new_song_accumulate))
```

```python
test_user_hist = test_data['msno'].value_counts()
print(test_user_hist)
```

```python
new_user_count = 0
new_user_accumulate = 0
less_user_count = 0
less_user_accumulate = 0
total_test_users = 0
total_test_user_accumulate = 0
for user_id, count_num in test_user_hist.iteritems():
    train_user_count = user_hist.get(user_id, default=-1)
    total_test_users += 1
    total_test_user_accumulate += count_num
    if train_user_count < 0:
        new_user_accumulate += count_num
        new_user_count += 1
    elif train_user_count < 5:
        less_user_count += 1
        less_user_accumulate += count_num
print('there are {}(total {}) users in test dataset'.format(total_test_users, total_test_user_accumulate))
print('{}/{} of them are count less than 5 in train dataset'.format(less_user_count, less_user_accumulate))
print('{}/{} of them never exists in train dataset'.format(new_user_count, new_user_accumulate))
```
